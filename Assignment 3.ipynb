```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("netflix_titles.csv")
```

```python
data.isnull().sum()
```

```python
#Making all NaN Values zero

data.replace(to_replace=np.nan,value=0,inplace=True)
data
```

```python
#to check

data.isnull().sum()
```

# Question 1



```python
#Adding a column Num_of_Cast

#defining a function

def set_Cast(val):
    if val == 0:
        return "none"
    else:
        return len(val.split(", "))

#Adding column

data["Num _of_Cast"] = data["cast"]. apply(set_Cast)
data
```
